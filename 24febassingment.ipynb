```python
import pandas as pd

def reindex_dataframe(df):
    new_index = pd.RangeIndex(start=1, step=2, stop=len(df) * 2 + 1)
    df = df.reset_index(drop=True)
    df.index = new_index
    return df
```

```python
df = pd.DataFrame({'A': [10, 20, 30], 'B': [40, 50, 60], 'C': [70, 80, 90]})
reindexed_df = reindex_dataframe(df)
print(reindexed_df)
```

```python
import pandas as pd

def cal_sum(df):
    sum=df['Values']
    col_sum=sum.head(3)
    first_three=col_sum.sum()
    print("sum of first three element :",first_three)
```

```python
df = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})
cal_sum(df)
```

```python
import pandas as pd

def add_word_count(df):
    df['Word_Count'] = df['Text'].apply(lambda x: len(str(x).split()))
    return df
```

```python
df = pd.DataFrame({'Text': ['This is a sample text', 'Another sentence', 'One more example']})
df = add_word_count(df)
print(df)
```
